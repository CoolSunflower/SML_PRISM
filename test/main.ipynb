```python
import pandas as pd
```

```python
df = pd.read_csv('HistoricData.csv')
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('portuguese'))
```

```python
for item in df[df['Subtopics'] == 'Fixos']['Mention Content']:
    # if last three characters are not '...'
    if not item.endswith('...') and not item == 'Deleted or protected mention':   
        # okay so the problem is that, mention content is actually portugese and fixos is a common word in portugese
        # so we need to find the common words in portugese and remove them from the mention content
        # now how to do this?
        # we can use a portugese stopwords list
        words = item.split()
        if not any(word.lower() in stop_words for word in words):
            print(item)
        # okay so its printing nothing now
        # the stop_words list is too broad, it even contains works like 'a' which is a big problem
        # we need to find a better way to filter out common words
        # maybe we can use a custom list of common words in portugese
        # or we can use a frequency analysis to find the most common words in the dataset
# lets try frequency analysis

from collections import Counter
all_words = []
for content in df[df['Subtopics'] == 'Fixos']['Mention Content']:
    if not content.endswith('...') and not content == 'Deleted or protected mention':
        all_words.extend(content.split())
```

```python
# print the 100 most common words
word_counts = Counter(all_words)
for item in word_counts.most_common(100):
    print(item)
```

```python
# print all fixos entry that do not contain the word 'que'
for item in df[df['Subtopics'] == 'Fixos']['Mention Content']:
    if 'milhões' not in item and not item.endswith('...') and not item == 'Deleted or protected mention':   
        # print(item)
        # is there some way to detect the laguage of the text programmatically?
        # we can use langdetect library
        from langdetect import detect
        if detect(item) == 'pt':
            print(item)
```
